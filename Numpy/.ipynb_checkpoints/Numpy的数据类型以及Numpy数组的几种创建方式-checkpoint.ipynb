# 2. 创建Numpy数组的几种方式

## 2.1 从python列表中创建数组

```python
import numpy as np
np.array([1,2,3,4,5])
```

```python
import numpy as np
# 指定数据类型
np.array([1,2,3,4,5], dtype='float32')
```

## 2.2 用Numpy内置方法创建更高效

```python
# 创建一个长度为10且元素的值均为0的数组
np.zeros(10, dtype=int)
```

```python
# 创建一个3x5的浮点型数组，数组元素的值都是1
np.ones((3, 5), dtype=float)
```

```python
# 创建一个3x5的浮点型数组，数组的值都是3.14
np.full((3, 5), 3.14, dtype=float)
```

```python
# 创建一个浮点型数组，数组是一个线性序列
# 从0开始，到20结束，步长为2
np.arange(0, 20, 2, dtype=float)
```

```python
# 创建一个7个元素的数组，这7个元素均匀的分布在[0,1]
# 每两个元素间的差都是一样的
a = np.linspace(0,1,7)
```

```python
# 创建一个3x3的、在[0,1)均匀分布的随机数组成的数组
np.random.random((3,3))
```

```python
# 创建一个3x3、均值为0、方差为1的正态分布的随机数组
np.random.normal(0, 1, (3,3))
```

```python
# 创建一个3x3、[0,10)区间的整数随机数组
np.random.randint(0, 10, (3,3))
```

```python
# 创建一个3x3的单位矩阵
np.eye(3,3)
```

```python
# 创建一个含有3个元素的整型数组,未初始化
np.empty(3)
```
